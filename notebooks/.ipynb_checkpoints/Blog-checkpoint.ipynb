# The Anglo-Saxon Exception: Inequality Across Developed Economies

## Introduction

This blog post examines trends in income inequality across developed economies 
since the 1980s, with a particular focus on whether English-speaking nations 
have followed a different path to their continental European counterparts.

We use data from the World Inequality Database (WID) and the OECD to compare 
Gini coefficients and income shares across a selection of countries including 
the US, UK, Germany, France, Sweden, and Canada.

```python
import pandas as pd

# Load the raw data
df = pd.read_csv('../data/raw/oecd_gini.csv')

# First look at the data
print(df.shape)
print(df.head())
print(df.columns.tolist())
```

```python
# Keep only the columns we need
df_clean = df[['REF_AREA', 'Reference area', 'TIME_PERIOD', 'OBS_VALUE']].copy()

# Rename columns to something cleaner
df_clean.columns = ['country_code', 'country', 'year', 'gini']

# Convert year and gini to numeric
df_clean['year'] = pd.to_numeric(df_clean['year'], errors='coerce')
df_clean['gini'] = pd.to_numeric(df_clean['gini'], errors='coerce')

# Drop any missing values
df_clean = df_clean.dropna()

# Check the result
print(df_clean.shape)
print(df_clean.head(10))
print(df_clean['country'].unique())
print(df_clean['year'].min(), "-", df_clean['year'].max())
```

```python
# Save cleaned data to processed folder
df_clean.to_csv('../data/processed/gini_clean.csv', index=False)
print("Cleaned data saved successfully")
```

```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# Define country groups
anglo_saxon = ['United States', 'United Kingdom', 'Canada', 'Australia']
continental = ['Germany', 'France', 'Sweden', 'Norway', 'Denmark']

# Define distinct colours for each country
anglo_colors = {
    'United States': '#D62728',    # red
    'United Kingdom': '#FF7F0E',   # orange
    'Canada': '#9467BD',           # purple
    'Australia': '#E377C2'         # pink
}

continental_colors = {
    'Germany': '#1F77B4',          # blue
    'France': '#2CA02C',           # green
    'Sweden': '#17BECF',           # cyan
    'Norway': '#BCBD22',           # yellow-green
    'Denmark': '#8C564B'           # brown
}

# Plot
fig, ax = plt.subplots(figsize=(12, 7))

for country in df_clean['country'].unique():
    data = df_clean[df_clean['country'] == country].sort_values('year')
    if country in anglo_saxon:
        color = anglo_colors[country]
        linestyle = '-'
        linewidth = 2.5
    else:
        color = continental_colors[country]
        linestyle = '--'
        linewidth = 1.8
    ax.plot(data['year'], data['gini'],
            label=country, color=color,
            linewidth=linewidth, linestyle=linestyle)

ax.set_title('Income Inequality Across Developed Economies (1980–2023)',
             fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Gini Coefficient (Disposable Income)', fontsize=12)
ax.legend(loc='upper left', fontsize=9)
ax.grid(True, alpha=0.3)

# Add legend markers explaining line styles
solid_line = mlines.Line2D([], [], color='black', linestyle='-', linewidth=2, label='Anglo-Saxon economies (solid)')
dashed_line = mlines.Line2D([], [], color='black', linestyle='--', linewidth=2, label='Continental/Nordic economies (dashed)')
ax.legend(loc='upper left', fontsize=9)
handles, labels = ax.get_legend_handles_labels()
handles += [solid_line, dashed_line]
labels += ['— Anglo-Saxon (solid)', '-- Continental/Nordic (dashed)']
ax.legend(handles=handles, labels=labels, loc='upper left', fontsize=8)

plt.tight_layout()
plt.savefig('../output/figures/gini_trends.png', dpi=150)
plt.show()
print("Chart saved!")
```

```python
# Calculate inequality change over time for each country
print("=== CHANGE IN GINI COEFFICIENT (earliest to latest available year) ===\n")

changes = []
for country in df_clean['country'].unique():
    country_data = df_clean[df_clean['country'] == country].sort_values('year')
    earliest_year = country_data['year'].min()
    latest_year = country_data['year'].max()
    earliest_gini = country_data[country_data['year'] == earliest_year]['gini'].values[0]
    latest_gini = country_data[country_data['year'] == latest_year]['gini'].values[0]
    change = latest_gini - earliest_gini
    pct_change = (change / earliest_gini) * 100
    group = 'Anglo-Saxon' if country in anglo_saxon else 'Continental/Nordic'
    changes.append({
        'Country': country,
        'Group': group,
        'Earliest Year': earliest_year,
        'Earliest Gini': round(earliest_gini, 3),
        'Latest Year': latest_year,
        'Latest Gini': round(latest_gini, 3),
        'Absolute Change': round(change, 3),
        'Percentage Change': round(pct_change, 1)
    })

df_changes = pd.DataFrame(changes).sort_values('Absolute Change', ascending=False)
df_changes = df_changes.set_index('Country')
print(df_changes.to_string())

# Group level summary
print("\n=== AVERAGE CHANGE BY GROUP ===\n")
group_changes = df_changes.groupby('Group')[['Absolute Change', 'Percentage Change']].mean().round(3)
print(group_changes)
```

```python
# Calculate average Gini for each group per year
df_clean['group'] = df_clean['country'].apply(
    lambda x: 'Anglo-Saxon' if x in anglo_saxon else 'Continental/Nordic'
)

group_avg = df_clean.groupby(['year', 'group'])['gini'].mean().reset_index()

anglo_data = group_avg[group_avg['group'] == 'Anglo-Saxon'].sort_values('year')
continental_data = group_avg[group_avg['group'] == 'Continental/Nordic'].sort_values('year')

# Merge to get only years where both groups have data
merged = anglo_data.merge(continental_data, on='year', suffixes=('_anglo', '_continental'))
merged['gap'] = merged['gini_anglo'] - merged['gini_continental']

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Top chart - use merged data so both series have same length
ax1.plot(merged['year'], merged['gini_anglo'],
         color='crimson', linewidth=2.5, label='Anglo-Saxon average')
ax1.plot(merged['year'], merged['gini_continental'],
         color='steelblue', linewidth=2.5, label='Continental/Nordic average')
ax1.fill_between(merged['year'], merged['gini_anglo'], merged['gini_continental'],
                 alpha=0.15, color='crimson')
ax1.set_title('Group Average Gini Coefficients Over Time', fontsize=13, fontweight='bold')
ax1.set_ylabel('Average Gini Coefficient', fontsize=11)
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

# Bottom chart - the gap over time
ax2.plot(merged['year'], merged['gap'], color='purple', linewidth=2.5)
ax2.axhline(y=merged['gap'].mean(), color='gray', linestyle='--', alpha=0.7, label='Average gap')
ax2.fill_between(merged['year'], merged['gap'], alpha=0.15, color='purple')
ax2.set_title('Inequality Gap: Anglo-Saxon minus Continental/Nordic', fontsize=13, fontweight='bold')
ax2.set_xlabel('Year', fontsize=11)
ax2.set_ylabel('Gini Gap', fontsize=11)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../output/figures/gini_gap.png', dpi=150)
plt.show()
print("Chart saved!")
```

```python
import wbdata
import datetime

# Define our countries
country_codes = ['USA', 'GBR', 'DEU', 'FRA', 'SWE', 'CAN', 'AUS', 'NOR', 'DNK']

# Define indicators we want from World Bank
indicators = {
    'NY.GDP.PCAP.KD': 'gdp_per_capita',      # GDP per capita (constant USD)
    'NE.TRD.GNFS.ZS': 'trade_openness',       # Trade as % of GDP
    'GC.XPN.TOTL.GD.ZS': 'govt_spending',     # Government expenditure as % of GDP
    'SL.UEM.TOTL.ZS': 'unemployment'          # Unemployment rate
}

# Fetch data
df_wb = wbdata.get_dataframe(indicators, country=country_codes)
df_wb = df_wb.reset_index()
df_wb.columns = ['country', 'date', 'gdp_per_capita', 'trade_openness', 'govt_spending', 'unemployment']

print(df_wb.shape)
print(df_wb.head())
```

```python
# Clean World Bank data
# Extract year as integer from date
df_wb['year'] = pd.to_numeric(df_wb['date'], errors='coerce').astype('Int64')
df_wb = df_wb.drop(columns=['date'])

# Drop rows with missing year or all NaN values
df_wb = df_wb.dropna(subset=['year'])
df_wb = df_wb[df_wb['year'] >= 1980]

# Map World Bank country names to match our Gini data
country_name_map = {
    'United States': 'United States',
    'United Kingdom': 'United Kingdom',
    'Germany': 'Germany',
    'France': 'France',
    'Sweden': 'Sweden',
    'Canada': 'Canada',
    'Australia': 'Australia',
    'Norway': 'Norway',
    'Denmark': 'Denmark'
}

# Merge with Gini data
df_merged = df_clean.merge(df_wb, on=['country', 'year'], how='inner')
df_merged = df_merged.dropna()

print(df_merged.shape)
print(df_merged.head())
print(df_merged['country'].value_counts())
```

```python
import statsmodels.api as sm

# Prepare regression variables
# We'll regress Gini on our economic variables plus an Anglo-Saxon dummy
df_reg = df_merged.copy()

# Create Anglo-Saxon dummy variable (1 = Anglo-Saxon, 0 = Continental/Nordic)
df_reg['anglo_dummy'] = df_reg['country'].apply(
    lambda x: 1 if x in anglo_saxon else 0
)

# Log transform GDP per capita (standard in economics)
import numpy as np
df_reg['log_gdp'] = np.log(df_reg['gdp_per_capita'])

# Define dependent and independent variables
y = df_reg['gini']
X = df_reg[['anglo_dummy', 'log_gdp', 'trade_openness', 'govt_spending', 'unemployment']]
X = sm.add_constant(X)

# Run OLS regression
model = sm.OLS(y, X).fit()

# Print results
print(model.summary())
```

```python
# Visualise regression coefficients
fig, ax = plt.subplots(figsize=(10, 6))

# Get coefficients and confidence intervals (excluding constant)
coefs = model.params[1:]
conf_int = model.conf_int().iloc[1:]

lower_err = coefs.values - conf_int.iloc[:, 0].values
upper_err = conf_int.iloc[:, 1].values - coefs.values

colors_coef = ['crimson' if c > 0 else 'steelblue' for c in coefs.values]

labels = ['Anglo-Saxon\nDummy', 'Log GDP\nper Capita', 
          'Trade\nOpenness', 'Government\nSpending', 'Unemployment']

ax.barh(labels, coefs.values, 
        xerr=[lower_err, upper_err],
        color=colors_coef, edgecolor='white', capsize=5)

ax.axvline(x=0, color='black', linewidth=0.8)
ax.set_title('OLS Regression: Determinants of Income Inequality\n(Dependent variable: Gini coefficient)', 
             fontsize=13, fontweight='bold')
ax.set_xlabel('Coefficient estimate (with 95% confidence interval)', fontsize=11)
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('../output/figures/regression_coefficients.png', dpi=150)
plt.show()
print("Chart saved!")
```

# The Anglo-Saxon Exception: Inequality Across Developed Economies

## Introduction

Why is it that people working the same job in London and Copenhagen live such different economic lives? The answer, at least in part, lies in how income is distributed unequally across society, and the data suggest that English-speaking economies have consistently followed a more unequal path than their continental European and Scandinavian peers.

To measure this, economists use the Gini coefficient, a number between 0 and 1 where 0 represents perfect equality (everyone earns the same) and 1 represents perfect inequality (one person holds all the wealth). Most developed nations sit between 0.25 and 0.45, but within that seemingly narrow range lie meaningful differences in living standards, social mobility, and economic opportunity.

Income inequality is not only an economic issue but a social and political one. High inequality has been linked to lower social mobility, poorer health outcomes, and rising political polarisation, making it one of the defining policy challenges of our time.

Using data from the OECD spanning four decades, this post examines whether the United States, United Kingdom, Canada and Australia have genuinely diverged from Germany, France and the Nordic economies on inequality, and what the data tells us about why.

## Data and Methodology 

The data used in this analysis comes from the OECD Income Distribution Database, one of the most comprehensive and consistently compiled sources of inequality data across developed economies, updated several times a year. 

Nine countries are compared: the United States, the United Kingdom, Australia, and Canada on one side, the four major Anglo-Saxon economies, and France, Germany, Sweden, Norway, and Denmark on the other. These two groups were chosen because they compare two recognised models of capitalism. The Anglo-Saxon economies are liberal market economies, whereas the continental/Nordic economies are coordinated or social market economies. The country selection also holds development constant; all countries are high-income countries, which helps to remove development as a variable, meaning that differences are likely from institutions and policy, rather than poverty. Together, these criteria ensure the selection is both comparable in terms of development and meaningfully varied in terms of institutions and policy, the exact variation this analysis seeks to explain. All nine countries also have consistent data availability across the full 1980-2023 period, ensuring the time-series comparisons are reliable. 

The measure used throughout is the disposable income Gini coefficient, that is, inequality measured after taxes and government transfers like benefits and pensions. This is the most meaningful measure for comparing what households actually have to live on, and it is standardised across OECD countries, making cross-country comparisons reliable. The data spans 1980 to 2023, allowing us to examine how inequality has evolved over four decades. The measure also reflects the effect of government policy, which is key when comparing countries with different redistribution policies.

## A Tale of Two Worlds

The divergence is clear and immediate. Since 1985, Anglo-Saxon economies have been more unequal than their continental and Nordic peers, and that gap has never closed. The United States leads the rankings throughout the entire period, reaching a Gini coefficient of 0.394 by 2023, compared to just 0.243 in Denmark, a difference of more than 60%.

The roots of this divergence can be linked to the political landscape of the 1980s. In the UK, Thatcher introduced strong restrictions on trade unions, reducing membership, leading to fewer collective bargaining agreements, meaning workers in declining industries saw wages stagnate. She also privatised many state-owned industries like British Telecom (BT), increasing competition and cost-cutting, increasing redundancies in lower-skilled jobs, while executives and shareholders saw rising pay and profits. In the US, Reagan introduced tax cuts for high-income earners, reducing the marginal income tax rate from 70% to 28%. He also famously fired striking air-traffic controllers and signalled for a tough stance on unions, decreasing union density and workers' bargaining power. Sectors like finance and energy were deregulated, and cuts were made to welfare programs. The overall impact of these two leaders was that wage inequality widened, financial sectors grew, salaries increased at the top of the income distribution, while tax reforms benefited high-income earners, and shrinking welfare provisions meant less redistribution, increasing disposable income inequality.

In contrast, the Nordic countries maintained strong welfare state expansion throughout the 1980s-90s, investing in universal healthcare, education, and unemployment insurance, while maintaining some of the world's highest tax rates and strongest union coverage. Crucially, they did not aggressively deregulate their financial sectors. 

Germany, at the same time, followed a coordinated market economy, characterised by strong worker representation, with powerful unions and workers on company boards, sector-based wage bargaining, where wages were set collectively by powerful industry unions. The country also maintained generous unemployment benefits, strong pensions schemes, keeping redistribution strong, and it did not pursue large-scale privatisation or tax cuts like the UK and the US. France also resisted the Anglo-Saxon model, with welfare expansion and high public spending, expanding healthcare, pensions, and unemployment insurance. They also maintained strong unions, high taxation and public sector dominance, reducing the profit-maximising incentive of private institutions. 

It is worth noting that inequality has risen across all nine countries since 1980; this is not a story of Anglo-Saxon decline against a stable European backdrop. But the starting points differ dramatically, and those differences have proven remarkably durable.

Looking at average Gini coefficients over the full 1990–2023 period makes the separation even cleaner. Every Anglo-Saxon economy sits above every Continental or Nordic economy without exception. Canada, at 0.307, sits closest to the European group, a reminder that the Anglo-Saxon label covers meaningful variation, but even Canada sits comfortably above France (0.289) and well above Denmark (0.243).

## Persistent but Not Widening

What is surprising about this analysis is that the inequality gap between Anglo-Saxon and Continental/Nordic economies remains stable. Figure 3 shows that the gap between the two groups' averages hovers around 0.07 Gini points for nearly four decades, rarely rising above 0.09 or falling below 0.05. Throughout the period, both groups become more unequal; that said, the Continental/Nordic group has relatively worsened, as seen in Table 1. 

As previously covered, inequality in Anglo-Saxon economies in the 1980s was far worse than that of their Continental/Nordic peers, but this low starting point means the Continental/Nordic group becomes relatively more unequal over time. A large part of this is driven by the erosion of the traditional welfare model that these economies employed. From the 1990s, many economies scaled back their high union coverage, compressed wage structures, and strong distribution to remain competitive in a more globalised economy. Many Nordic economies also began reducing their high top marginal tax rates, shifting to more consumption taxes, impacting the redistributive impact of the tax system. Also, many Nordic and Continental economies saw large inflows of migrants in the 1990s-2010s. Combined with strict employment protection, this produced dual labour markets, more low-pay service jobs at the bottom, and rising inequality within native workers and between natives and migrants. The increase in low-paying service jobs is especially compelling as Nordic economies have been becoming more dependent on knowledge-intensive, high-tech sectors, where capital incomes rose faster than wages, which were disproportionately held by top earners. 

Sweden specifically saw the sharpest relative increase in inequality, a percentage change of 46%, moving from the world's most equal country to a more liberalised model. Sweden's reforms in the 90s saw significant cuts to top marginal tax rates, reductions in capital taxation, and some universal benefits replaced with means-tested welfare. Technological change also hit Sweden hard. Sweden is now a global leader in ICT and high-tech, being home to companies like Ericsson, Spotify, and Klarna, where highly skill-based sectors drive fast growth in top incomes and stronger urban wage premiums (Stockholm vs. the rest of the country). Sweden also saw higher immigration rates than almost any European country relative to population size. Combining all this with Sweden's very equal starting point explains why it saw the highest relative increase of all countries examined. 

Anglo-Saxon economies saw less growth in inequality relatively, largely due to them already having weaker unions, flatter wage structures, and less redistributive welfare states. Australia especially stands out as it maintained high minimum wages, preserved sectoral wage awards, and retained a relatively progressive fiscal system. This can explain why they only saw a 3.2% increase in inequality across the examined period. Effectively, many of the reasons that Anglo-Saxon economies were more unequal in the 1980s (liberalised economies, less powerful unions, less redistribution) explain why they don't see as much of an increase across the period. Many of the policies that the Continental/Nordic group implemented in the 1990s-2000s had already been implemented by the Anglo-Saxon economies in the 1980s. 

A notable feature is the sharp dip in Anglo-Saxon inequality around 2020, driven largely by emergency stimulus payments deployed by the US, UK, Canadian and Australian governments during the Covid-19 pandemic. These transfers temporarily boosted incomes at the bottom of the distribution, compressing measured inequality, but the effect proved short-lived, with inequality rebounding sharply by 2022 as support programmes were withdrawn. While the Continental/Nordic group had some form of stimulus programme during the pandemic, this was more in the form of short-time work schemes, where governments would pay a portion of workers' wages, as opposed to cash transfers like in the Anglo-Saxon group.

## What Drives the Anglo-Saxon Exception?

The regression analysis gets to the core of the question this post set out to answer. Even after controlling for GDP per capita, trade openness, government spending and unemployment, Anglo-Saxon economies have a Gini coefficient around 0.063 higher than their Continental and Nordic peers. This coefficient is highly statistically significant, and the model as a whole explains 72% of the variation in Gini coefficients across the sample, a strong result for cross-country economic data. 

The positive Log GDP per capita coefficient suggests a possible positive relationship between wealth and inequality, though this does not reach statistical significance in this sample and should be interpreted cautiously. This pattern, if real, would be consistent with skill-biased tech changes, larger returns to capital in advanced economies and rising top-income shares in high-income countries, though the lack of significance means this interpretation remains tentative. 

The other variables included have a small coefficient size, indicating that they have little effect on inequality within the examined economies. Interestingly, the government spending variable has a positive impact on inequality. This does not mean government spending increases inequality. Rather, it likely reflects reverse causality; countries with higher inequality tend to spend more on redistribution in response to that inequality, but this spending is not always sufficient to fully close the gap. It is a reminder that redistribution treats the symptom rather than the cause.

What this means is that the Anglo-Saxon inequality premium cannot be explained by standard economic variables. The countries aren't more unequal because they are richer, more exposed to global trade, or facing higher unemployment. More fundamental institutional differences appear to be driving the divergence: welfare-state design, labour-market regulation, union strength, minimum wages, and tax progressivity. The growing financialisation of Anglo-Saxon economies, where financial services account for a disproportionate share of output and executive compensation, has widened the gap between top earners and the rest.

However, there are some limitations to this regression. The regression only includes a small set of macroeconomic controls plus a group dummy. This means that many crucial drivers of inequality remain unobserved or are impossible to quantify cleanly. Many of the most important determinants of inequality are institutional, and these do not fit neatly into a regression. Factors like labour-market regulation, tax progressivity, and welfare-state architecture cannot be measured by the regression. Many historical and political factors also cannot be included, so while the regression confirms the existence of the Anglo-Saxon premium, it cannot tell us precisely what causes it.

## Conclusion

Four decades of data tell a consistent and striking story. English-speaking economies are more unequal than their continental European and Nordic peers. The gap between the two groups hasn't changed significantly over the period, remaining around 0.07 Gini points. This can be explained by the convergence paradox, where the Continental/Nordic group has relatively become far more unequal across the period, rising by around 24% compared to the Anglo-Saxon group's 8.7%. The Continental/Nordic group started less unequal, but over time, eroded their welfare models due to globalisation pressures, tax reform, migration and dual labour markets. The absolute numbers make this clear. The United States is the most unequal, with a final Gini score of 0.394, more than 60% higher than Denmark's 0.243. The separation holds across every country and almost every year in the sample. If current trends continue, the question is not whether Anglo-Saxon economies will become more equal, but whether Continental/Nordic economies will eventually converge upward toward Anglo-Saxon levels of inequality rather than the reverse. And if they do, will that rising inequality introduce the dynamic labour markets and innovation environments that Anglo-Saxon proponents argue justify the trade-off, or will it simply be inequality for inequality's sake, without the compensating economic dynamism?

What is surprising is what the regression reveals. After accounting for differences in GDP, trade exposure, government spending and unemployment, Anglo-Saxon economies carry an inequality premium of around 0.063 Gini points. Standard economic variables cannot explain this gap. The origins of the Anglo-Saxon exception seem to come from historical factors like the political choices of Thatcher and Reagan in the 1980s, which saw declining trade union power, financialisation of the economies, and welfare state models that redistribute income without addressing the drivers of wage inequality (the symptom is being treated, not the cause). One thing worth keeping in mind while interpreting this regression is that the analysis includes only 9 countries and a limited set of variables. A richer dataset with more countries and institutional variables like union density, replacement rates, and top marginal tax rates would allow us to make stronger causal claims. 

Whether the gap is inevitable is an open question. Some economists believe it reflects a trade-off, that the flexible labour markets of Anglo-Saxon economies create stronger growth and innovation at the cost of higher inequality. Others believe it reflects political choices that could be reversed through stronger labour market institutions, higher minimum wages, and more progressive taxation. What the data cannot tell us is which of these interpretations is correct, but it does tell us, clearly and consistently, that the gap is real, it is large, and four decades on, it shows no sign of closing.

For the worker in Copenhagen, that difference still matters enormously. But if current trends continue, it may matter rather less to their children.
